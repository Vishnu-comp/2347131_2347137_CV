```python
import torch
import cv2
import numpy as np

# Download YOLOv5 model (pre-trained)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # 'yolov5s' is a small and fast model

# Define a function for Student Attendance Monitoring  and counting
def detect_and_count_people(frame):
    # Convert BGR (OpenCV default) to RGB (required by YOLOv5)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Perform inference
    results = model(frame_rgb)

    # Parse results
    detections = results.pandas().xyxy[0]  # Pandas DataFrame with detection results

    person_count = 0
    for _, row in detections.iterrows():
        if row['name'] == 'person':  # Only count people
            x_min, y_min, x_max, y_max = int(row['xmin']), int(row['ymin']), int(row['xmax']), int(row['ymax'])
            confidence = row['confidence']

            # Draw bounding box
            cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

            # Put label and confidence
            label = f"Person {confidence:.2f}"
            cv2.putText(frame, label, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

            person_count += 1

    # Display the count on the frame
    cv2.putText(frame, f"Count: {person_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    return frame, person_count

# Test the function
if __name__ == "__main__":
    # Load the image
    image_path = "stud.jpg"  # Replace with the path to your image file
    frame = cv2.imread(image_path)

    if frame is None:
        print("Error: Could not load image.")
        exit()

    # Perform Student Attendance Monitoring  and counting
    frame, person_count = detect_and_count_people(frame)

    # Get the image dimensions
    height, width, _ = frame.shape

    # Create a window size based on the image dimensions to avoid zooming
    window_name = "Student Attendance Monitoring "
    cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)
    cv2.resizeWindow(window_name, width, height)

    # Display the result
    cv2.imshow(window_name, frame)

    # Wait for any key to close the image window
    cv2.waitKey(0)

    # Release resources
    cv2.destroyAllWindows()
```

```python
import torch
import cv2
import numpy as np
from tkinter import Tk, filedialog, Button, Label, Frame
from PIL import Image, ImageTk
from tkinter import font as tkfont

# Download YOLOv5 model (pre-trained)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # 'yolov5s' is a small and fast model

# Define a function for ATTENDANCE COUNT MANAGEMENT and counting
def detect_and_count_people(frame):
    # Convert BGR (OpenCV default) to RGB (required by YOLOv5)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Perform inference
    results = model(frame_rgb)

    # Parse results
    detections = results.pandas().xyxy[0]  # Pandas DataFrame with detection results

    person_count = 0
    for _, row in detections.iterrows():
        if row['name'] == 'person':  # Only count people
            x_min, y_min, x_max, y_max = int(row['xmin']), int(row['ymin']), int(row['xmax']), int(row['ymax'])
            confidence = row['confidence']

            # Draw bounding box
            cv2.rectangle(frame, (x_min, y_min), (x_max, y_max), (0, 255, 0), 2)

            # Put label and confidence
            label = f"Person {confidence:.2f}"
            cv2.putText(frame, label, (x_min, y_min - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

            person_count += 1

    # Display the count on the frame
    cv2.putText(frame, f"Count: {person_count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

    return frame, person_count

# Function to load and process image
def load_image():
    file_path = filedialog.askopenfilename(title="Select an Image", filetypes=[("Image files", "*.jpg *.jpeg *.png")])
    if file_path:
        frame = cv2.imread(file_path)
        if frame is None:
            print("Error: Could not load image.")
            return
        frame, person_count = detect_and_count_people(frame)
        
        # Convert to ImageTk format for displaying in Tkinter
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img = Image.fromarray(frame_rgb)

        # Resize the image to fit the Tkinter window
        img = resize_image(img)

        img = ImageTk.PhotoImage(img)
        
        panel.config(image=img)
        panel.image = img

# Function to load and process video
def load_video():
    file_path = filedialog.askopenfilename(title="Select a Video", filetypes=[("Video files", "*.mp4 *.avi")])
    if file_path:
        cap = cv2.VideoCapture(file_path)
        if not cap.isOpened():
            print("Error: Could not open video.")
            return
        
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            frame, person_count = detect_and_count_people(frame)
            
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(frame_rgb)

            # Resize the image to fit the Tkinter window
            img = resize_image(img)

            img = ImageTk.PhotoImage(img)
            
            panel.config(image=img)
            panel.image = img
            
            root.update()  # Update the Tkinter window
            # Exit on key press
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
        
        cap.release()

# Function to start webcam capture
def start_webcam():
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Error: Could not open webcam.")
        return

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame, person_count = detect_and_count_people(frame)
        
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img = Image.fromarray(frame_rgb)

        # Resize the image to fit the Tkinter window
        img = resize_image(img)

        img = ImageTk.PhotoImage(img)
        
        panel.config(image=img)
        panel.image = img
        
        root.update()  # Update the Tkinter window
        
        # Exit on key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    
    cap.release()

# Function to resize image to fit the Tkinter window
def resize_image(img):
    # Get the current size of the Tkinter window
    window_width = root.winfo_width()
    window_height = root.winfo_height()

    # Resize the image to fit the window (preserving the aspect ratio)
    img.thumbnail((window_width, window_height), Image.ANTIALIAS)

    return img

# Create the main window
root = Tk()
root.title("ATTENDANCE COUNT MANAGEMENT with YOLOv5")

# Set the size of the window
root.geometry("900x700")

# Set the background color to a subtle gradient
root.config(bg="#f5f5f5")

# Create a gradient header
header_frame = Frame(root, bg='#2c3e50', padx=20, pady=10)
header_frame.pack(fill="x", side="top", anchor="n")

header_label = Label(header_frame, text="Attendance Count Management", fg='white', bg='#2c3e50', font=("Helvetica", 18, "bold"))
header_label.pack()

# Create a frame for buttons
button_frame = Frame(root, bg='#f5f5f5', pady=20)
button_frame.pack()

# Define custom font style for buttons
custom_font = tkfont.Font(family="Helvetica", size=14, weight="bold")

# Create buttons with hover effects, rounded corners, and tooltips
def on_enter(e):
    e.widget['bg'] = '#27ae60'
    
def on_leave(e):
    e.widget['bg'] = '#4CAF50'

btn_image = Button(button_frame, text="Load Image", command=load_image, font=custom_font, bg='#4CAF50', fg='white', relief="flat", width=20, height=2, bd=0, borderwidth=0)
btn_image.pack(pady=10)
btn_image.bind("<Enter>", on_enter)
btn_image.bind("<Leave>", on_leave)

btn_video = Button(button_frame, text="Load Video", command=load_video, font=custom_font, bg='#4CAF50', fg='white', relief="flat", width=20, height=2, bd=0, borderwidth=0)
btn_video.pack(pady=10)
btn_video.bind("<Enter>", on_enter)
btn_video.bind("<Leave>", on_leave)

btn_webcam = Button(button_frame, text="Start Webcam", command=start_webcam, font=custom_font, bg='#4CAF50', fg='white', relief="flat", width=20, height=2, bd=0, borderwidth=0)
btn_webcam.pack(pady=10)
btn_webcam.bind("<Enter>", on_enter)
btn_webcam.bind("<Leave>", on_leave)

# Create a panel to display images or video
panel = Label(root)
panel.pack(fill="both", expand=True)

# Run the application
root.mainloop()
```
